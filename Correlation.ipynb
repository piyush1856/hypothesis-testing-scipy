````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import pearsonr, spearmanr
````

````python
!wget --no-check-certificate https://drive.google.com/uc?id=1M6ICGsUhO-IIFl_HPYpk9vJO-GJEszmC -O weight-height.csv
````

````python
df_hw = pd.read_csv("weight-height.csv")
df_hw.head()
````

````python
sns.scatterplot(x=df_hw["Height"], y=df_hw["Weight"])
````

````python
df_hw[["Height", "Weight"]].corr()
````

````python
# Taking the above Rank(Experience) as x and Rank(Salary) as y

x = [1, 3, 4, 5, 2]
y = [1, 2, 5, 4, 3]

# calculate Spearman's correlation coefficient and p-value
corr, pval = spearmanr(x, y)

# print the result
print("Spearman's correlation coefficient:", corr)
````

````python
# H0: No correlation
# Ha: There is correlation
pearsonr(df_hw["Height"], df_hw["Weight"])
````

````python
# H0: No correlation
# Ha: There is correlation
spearmanr(df_hw["Height"], df_hw["Weight"])
````

````python
# sample data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

# calculate Spearman's correlation coefficient and p-value
corr, pval = spearmanr(x, y)

# print the result
print("Spearman's correlation coefficient:", corr)
print("p-value:", pval)
````
